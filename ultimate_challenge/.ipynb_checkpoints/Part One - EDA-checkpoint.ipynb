```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
logins = pd.read_json('logins.json')
```

```python
logins.head()
```

```python
logins['count'] = 1
logins.set_index(pd.DatetimeIndex(logins['login_time']), inplace=True)
min15 = logins.resample('15T').sum()
```

```python
min15.head()
```

```python
min15.index.year.value_counts()
```

All of these are dated 1970. Based on the other dataset, I suspect the year is supposed to be 2014. 

```python
min15['count'].hist(bins=20)
plt.title('Frequency of Logins in each 15 minute period')
plt.xlabel('Logins Per 15 Minute Period')
plt.ylabel('Number of Time Periods')
plt.show()
```

```python
plt.figure(figsize=(15,7))
min15.resample('D')['count'].sum().plot()
plt.xlabel('Date')
plt.show()
```

There appears to be a weekly cycle to the number of logins, and the peaks of the logins are generally increasing over time indicating a growing number of active users.

```python
min15['day'] = min15.index.dayofweek
```

```python
hourly = logins.resample('H').sum()
hourly.groupby(hourly.index.hour)['count'].mean().plot(kind='bar')
plt.title('Average Logins by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Avg # of Logins')
plt.show()
```

The least active hours seem to be 6, 7, and 8 am. How do these numbers vary by day of the week. 

```python
min15['count'].hist(bins=20)
plt.title('Frequency of Logins in each 15 minute period')
```

```python
min15.drop(columns='hour').groupby('day').sum().plot(kind='line')
plt.xlabel('Day')
plt.xticks(range(0,7), labels = ['M','T','W','Th','F','S','S'])
plt.title('Average Number of Logins Per Day')
plt.show()
```

```python
min15['hour'] = min15.index.hour
g = sns.FacetGrid(data=min15, col='hour', col_wrap = 3)
g.map(sns.barplot, 'day', 'count', order=range(0,7))
g.set_titles("{col_name}" + ":00")
g.set_axis_labels(x_var="Day of Week", y_var="Avg Logins")
for ax in g.axes.flat:
    ax.set_xticklabels(['M','T','W','Th','F','S','S'])
plt.show()
```

Saturdays and Sundays have the highest rate of logins between the hours of 1 AM and 7 AM.
In general, weekends see more logins than weekdays with the exception of the hours between 9 AM and 1 PM. 
Friday nights see the more logins than the weekend between 8 pm and midnight. 

```python
pivot15 = min15
pivot15 = pivot15.groupby(['day','hour']).mean().reset_index()
sns.relplot(data=pivot15, x='hour', y = 'count', palette = 'PuOr',
            linewidth=2, hue='day', kind='line', aspect=2, legend=False)
plt.legend(labels=['M','T','W','Th','F','S','S'])
plt.title('Hourly Logins by Day of Week')
plt.xlabel('Hour')
plt.ylabel('Logins')
plt.show()
```
